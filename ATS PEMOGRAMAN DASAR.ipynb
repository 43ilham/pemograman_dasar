```python
Penjelasan singkat:

Operator and digunakan karena kedua kondisi harus terpenuhi.

Kondisi pertama: tekanan_uap > 150

Kondisi kedua: status_manual_override == False

Jika keduanya benar, maka program mencetak "Katup terbuka", selain itu "Katup tertutup".
```

```python
def sistem_katup(tekanan_uap, status_manual_override):
    """
    Fungsi ini mensimulasikan logika kerja katup pengaman (safety valve).

    Parameter:
        tekanan_uap (float): nilai tekanan uap (dalam psi)
        status_manual_override (bool): status override manual
                                       True  = katup dikunci manual (tidak boleh terbuka)
                                       False = sistem otomatis aktif (boleh terbuka)

    Return:
        str: status katup ('Terbuka' atau 'Tertutup')
    """

    print("=== SISTEM PEMANTAUAN KATUP PENGAMAN ===")
    print(f"Tekanan Uap Saat Ini     : {tekanan_uap} psi")
    print(f"Status Manual Override   : {status_manual_override}")
    print("-------------------------------------------")

    if tekanan_uap > 150 and status_manual_override == False:
        print("Kondisi terpenuhi:")
        print(" - Tekanan melebihi ambang batas 150 psi")
        print(" - Override manual TIDAK aktif (False)")
        print(">>> Hasil: Katup Terbuka untuk melepaskan tekanan berlebih.")
        status_katup = "Terbuka"
    else:
        print("Kondisi tidak terpenuhi:")
        if tekanan_uap <= 150:
            print(" - Tekanan masih di bawah batas aman (<= 150 psi).")
        if status_manual_override == True:
            print(" - Override manual aktif, sistem dikunci oleh operator.")
        print(">>> Hasil: Katup Tetap Tertutup demi keamanan.")
        status_katup = "Tertutup"

    print("===========================================\n")
    return status_katup


hasil1 = sistem_katup(160, False)

hasil2 = sistem_katup(120, False)

hasil3 = sistem_katup(180, True)

print("=== RANGKUMAN HASIL PENGUJIAN ===")
print(f"Skenario 1 (160 psi, override False): Katup {hasil1}")
print(f"Skenario 2 (120 psi, override False): Katup {hasil2}")
print(f"Skenario 3 (180 psi, override True ): Katup {hasil3}")
print("===================================")
```

```python
Penjelasan Lengkap:

def sistem_katup(...): Fungsi utama untuk mensimulasikan sistem katup pengaman.

Docstring ("""...""") Memberi dokumentasi tentang parameter dan fungsinya, agar mudah dipahami oleh teknisi lain.

Bagian print(...) di awal fungsi Menampilkan data kondisi nyata dari sistem sebelum keputusan dibuat:

Tekanan uap saat ini

Status override manual

if tekanan_uap > 150 and status_manual_override == False: Kondisi majemuk menggunakan operator AND. Kedua kondisi harus benar agar katup boleh terbuka.

Bagian else: Menangani semua kondisi selain itu â€” misalnya tekanan belum cukup tinggi, atau override aktif.

Percabangan tambahan di dalam else: Memberi tahu alasan spesifik mengapa katup tidak terbuka:

Tekanan belum cukup tinggi

Override manual aktif

Tiga contoh pengujian (test case):

Tekanan tinggi, override nonaktif â†’ katup terbuka

Tekanan rendah, override nonaktif â†’ tetap tertutup

Tekanan tinggi, override aktif â†’ tertutup
```

```python
#Challenge
def hitung_daya(volt, arus):
    """Menghitung daya listrik (Watt)"""
    return volt * arus

def hitung_biaya_listrik(kwh):
    """Menghitung biaya listrik berdasarkan penggunaan"""
    tarif = 1500
    return kwh * tarif

def simulasi_penggunaan():
    print("=== Simulasi Energi Listrik Rumah Tangga ===")
    total_kwh = 0


    for i in range(1, 9):
        print(f"\nPeralatan ke-{i}")
        nama = input("Masukkan nama peralatan: ")
        daya = float(input("Masukkan daya peralatan (Watt): "))
        waktu = float(input("Lama pemakaian (jam/hari): "))

        energi = (daya * waktu) / 1000
        print(f"Energi harian {nama}: {energi:.2f} kWh")
        total_kwh += energi


    print("\n=== Hasil Simulasi ===")
    print(f"Total penggunaan energi harian: {total_kwh:.2f} kWh")

    if total_kwh < 5:
        print("Konsumsi energi rendah âš¡")
    elif total_kwh <= 15:
        print("Konsumsi energi sedang âš™ï¸")
    else:
        print("Konsumsi energi tinggi! ðŸ”¥")

    biaya = hitung_biaya_listrik(total_kwh)
    print(f"Perkiraan biaya listrik harian: Rp{biaya:,.0f}")

while True:
    simulasi_penggunaan()
    ulang = input("\nIngin simulasi lagi? (y/n): ").lower()
    if ulang != 'y':
        print("\nProgram selesai. Terima kasih telah menggunakan simulasi energi listrik âš¡")
        break
```

```python
#SOAL NO 2
def hitung_efisiensi(input_energi, output_energi):
    """
    Fungsi ini menghitung efisiensi turbin.
    Parameter:
        input_energi (float): Energi yang masuk ke sistem (Joule, kJ, dll)
        output_energi (float): Energi keluaran dari sistem (Joule, kJ, dll)
    Return:
        float: Nilai efisiensi (antara 0 - 1)
    """

    if input_energi == 0:
        print("Error: Input energi tidak boleh nol.")
        return None

    efisiensi = output_energi / input_energi

    print("=== Perhitungan Efisiensi Turbin ===")
    print(f"Input Energi   : {input_energi}")
    print(f"Output Energi  : {output_energi}")
    print(f"Efisiensi Turbin: {efisiensi:.2f} atau {efisiensi * 100:.1f}%")

    return efisiensi


hasil = hitung_efisiensi(500, 400)

if hasil is not None:
    print(f"\nNilai efisiensi yang dihitung adalah {hasil:.2f} atau {hasil * 100:.1f}%")
```
